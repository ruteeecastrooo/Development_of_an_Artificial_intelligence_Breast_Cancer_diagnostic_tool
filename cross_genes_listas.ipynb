```python
#hascaner part
# Lê os arquivos de texto e armazena os genes em listas
with open('sorted_f_value_hascancer.txt', 'r') as f:
    genes_1500 = f.read().splitlines()

with open('pam50.txt', 'r') as f:
    genes_50 = f.read().splitlines()

# Inicializa uma lista vazia para armazenar os genes correspondentes e seus valores
matching_genes = []

# Percorre cada gene na lista menor e verifica se é uma substring de algum gene na lista maior
for gene_small in genes_50:
    for gene_entry in genes_1500:
        gene_large, value_str = gene_entry.split("\t")  # Assumindo que o nome do gene e o valor são separados por um tab
        value = float(value_str)  # Converte o valor para float
        if gene_small in gene_large:
            matching_genes.append((gene_small, gene_large, value))

# Ordena os genes correspondentes com base no valor, do maior para o menor
matching_genes.sort(key=lambda x: x[2], reverse=True)

# Inicializa a string que será escrita no arquivo de texto
output_str = ""

# Imprime os genes correspondentes e adiciona à string de saída
if matching_genes:
    output_str += "Genes correspondentes encontrados:\n"
    for small, large, value in matching_genes:
        output_str += f"{small} corresponde a {large} com valor {value}\n"
    output_str += f"Total de genes correspondentes: {len(matching_genes)}\n"  # Adiciona a soma total de genes correspondentes
else:
    output_str += "Nenhum gene correspondente encontrado.\n"

print(output_str)

# Salva num txt
with open('matching_genes_results_hascancer.txt', 'w') as f:
    f.write(output_str)
```
